```python
# Celda 1: Importar librerías
import os
from dotenv import load_dotenv
from pypdf import PdfReader
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
# Celda 2: Probar carga de PDF
load_dotenv()
lector = PdfReader("../data/documento.pdf")
documentos = []
for i, pagina in enumerate(lector.pages):
        texto = pagina.extract_text()
        if texto.strip():
            documentos.append(Document(page_content=texto, metadata={"source": "../data/documento.pdf", "page": i}))
print(f"📄 Páginas: {len(documentos)}")
```

```python
# Celda 3: Ver contenido de una página específica
print(documentos[0].page_content[:500])

```

```python
# Celda 4: Probar división
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
fragmentos = splitter.split_documents(documentos)
print(f"📝 Fragmentos: {len(fragmentos)}")
```

```python

# Celda 5: Ver fragmentos generados
for i, frag in enumerate(fragmentos[:3]):
    print(f"\n--- Fragmento {i} ---")
    print(frag.page_content[:200])
```
